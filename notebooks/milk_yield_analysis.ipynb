```python
import pandas as pd
import glob

# Load all CSV files in data/ folder
files = glob.glob("../data/daily_cow_records_*.csv")

dfs = []
for f in files:
    df = pd.read_csv(f)
    df["source_file"] = f  # keep track of file origin
    dfs.append(df)

# Combine into one dataframe
data = pd.concat(dfs, ignore_index=True)

data.head()
```

```python
import pandas as pd
import glob

# Path to your data folder
path = "../data/*.csv"

# Load all CSVs
files = glob.glob(path)

files
```

```python
# Load and combine all CSVs into one dataframe
df_list = []

for file in files:
    temp_df = pd.read_csv(file)
    temp_df["source_file"] = file  # track origin
    df_list.append(temp_df)

df = pd.concat(df_list, ignore_index=True)

df.head()
```

```python
df.columns = (
    df.columns
    .str.replace("+AC0-", "-", regex=False)
    .str.replace("+AF8-", "_", regex=False)
    .str.replace("+", "", regex=False)
)

df.head()
```

```python
df['Date'] = df['Date'].str.replace("+AC0-", "-", regex=False)
df.head()
```
